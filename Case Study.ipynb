```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_excel('current2403.xlsx',index_col=0)
```

```python
df.shape
```

```python
df.tail(4)
```

```python
d=df[(df['price']>=2500000)&(df['price']<=4000000)]
d.shape
```

```python
d1=d[(d['place']=='Gachibowli')|(d['place']=='Miyapur')|(d['place']=='Kukatpally')]
d1.shape
```

```python
d2=d1['title'].value_counts()
d2
```

```python
d1.set_index(keys='title',inplace=True)
```

```python
d1.head(5)
```

```python
d3=d1.filter(like='Apartment',axis=0)
d3.head(5)
```

```python
d3.reset_index(inplace=True)
d3.head()
```

```python
# Returns no. of rows and columns  after filtering of the dataset

d3.shape
```

```python
d3.describe()
```

```python
d4=d3['title'].value_counts()
d4
```

```python
d4.plot.bar()
```

```python
d5=d3['place'].value_counts()
d5
```

```python
d5.plot.bar()
```

```python
explode=(0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05) # Distance between each sub-pie.
d3['price'].value_counts(ascending=False).head(10).plot(kind='pie',autopct="%0.2f%%",explode=explode)
```

```python
f,ax=plt.subplots(figsize=(15,10)) # Plot size
ax2=plt.subplot(223)
d3['price'].plot.hist() # Histogram for all prices.
ax2.plot

ax3=plt.subplot(224)
sns.distplot(d3['price'],ax=ax3) # Distribution plot for all prices.
```

```python
f,ax=plt.subplots(1,2,figsize=(15,5))
f.suptitle("Frequencies of rate_persqft and area_insqft") # Plot title
d3['per_sqft'].plot(kind='hist',ax=ax[0]) # Histogram for rate_persqft.

d3['area'].plot(kind='hist',ax=ax[1]) # Histogram for area_insqft.
```

```python
d3['status'].value_counts()
```

```python
fig,ax=plt.subplots(figsize=(8,5))
sns.countplot(d3['status'],ax=ax) # Count plot
plt.title("status of property") # Plot title
```

```python
d4=d3[(d3['area']>=1000)]
d4.shape
```

```python
d5=d4[(d4['title']=='2  BHK  Apartment')|(d4['title']=='3  BHK  Apartment')]
d5.shape
```

```python
d5['title'].value_counts()
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.boxplot(y='price',x='title',data=d5,ax=ax)
ax.set_xticklabels(ax.get_xticklabels(),rotation=50,ha='right')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sns.boxplot(y='price',x='place',data=d5,ax=ax)
ax.set_xticklabels(ax.get_xticklabels(),rotation=50,ha='right')
plt.show()
```

```python
sns.jointplot(x="area", y="per_sqft", data=d5)
```

```python
sns.catplot(x="place", y="price", hue="status",split=True, kind="violin", data=d5);
```
