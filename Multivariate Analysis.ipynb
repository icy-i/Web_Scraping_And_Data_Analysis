```python
c=df[(df['price']>=16000)&(df['price']<=1000000)]
d6_sort=c.sort_values(by='price',axis=0,ascending=False)
```

```python
def multi_scatter(col1,col2,col3,x,y):
    c=df[(df['price']>=x)&(df['price']<=y)]
    d6_sort=c.sort_values(by='price',axis=0,ascending=False)
    sns.set_style("whitegrid")
    ax1=sns.FacetGrid(d6_sort.head(30),hue=col3,height=5,aspect=2).map(plt.scatter,col1,col2,marker='+').set(yscale='log').add_legend()
    ax1.set_xticklabels(rotation=60)
    d6_sort=c.sort_values(by='price',axis=0,ascending=True)
    sns.set_style("whitegrid")
    ax2=sns.FacetGrid(d6_sort.head(30),hue=col3,height=5,aspect=2).map(plt.scatter,col1,col2,marker='+').set(yscale='log').add_legend()
    ax2.set_xticklabels(rotation=60)
    plt.show()
    
```

```python
multi_scatter('place','price','status',16000,1000000)
```

```python
multi_scatter('place','price','title',16000,1000000)
```

```python
def multi_pair(x,y,d):
    c=df[(df['price']>=x)&(df['price']<=y)]
    sns.pairplot(d6_sort,kind='reg')
```
